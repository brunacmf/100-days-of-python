```python
# set the logos
logo = """
    __  ___       __             
   / / / (_)___ _/ /_  ___  _____
  / /_/ / / __ `/ __ \/ _ \/ ___/
 / __  / / /_/ / / / /  __/ /    
/_/ ///_/\__, /_/ /_/\___/_/     
   / /  /____/_      _____  _____
  / /   / __ \ | /| / / _ \/ ___/
 / /___/ /_/ / |/ |/ /  __/ /    
/_____/\____/|__/|__/\___/_/     
"""

vs = """
 _    __    
| |  / /____
| | / / ___/
| |/ (__  ) 
|___/____(_)
"""
```

```python
# def a function to get a random account

def instagram_account():
  return random.choice(data)
```

```python
#def a function to properly print the values of the dicts

def print_data(account):
  return f" {account['name']}, a {account['description']}, from {account['country']}"
```

```python
#def a function to check the answer

def check(guess, a_followers, b_followers):
  if a_followers > b_followers:
    return guess == "a"
  else:
    return guess == "b"
```

```python
#def a function to play the game

def game():
  print(logo)
  score = 0
  game_should_continue = True
  a = instagram_account()
  b = instagram_account()

  while game_should_continue:
    a = b
    b = instagram_account()

    while a == b:
      b = instagram_account()

    print(f"Compare A: {print_data(a)}.")
    print(vs)
    print(f"Against B: {print_data(b)}.")
    
    guess = input("Who has more followers? Type 'A' or 'B': ").lower()
    a_follower_count = a["follower_count"]
    b_follower_count = b["follower_count"]
    is_correct = check(guess, a_follower_count, b_follower_count)
    
    print(logo)
    if is_correct:
      score += 1
      print(f"You're right! Current score: {score}.")
    else:
      game_should_continue = False
      print(f"Sorry, that's wrong. Final score: {score}")

```

```python
#call the game function

game()
```
